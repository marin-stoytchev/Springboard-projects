**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# drop null values and check by calling .info(); 
# since this is the dataframe we will be working with from now on create new data frame df

df = properties.dropna()
df.reset_index(drop = True, inplace = True)

print(df.info())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# create new data drame, dfg, where data is grouped by year and the prices are aggregated using mean

dfg = df.groupby(by=['Borough', 'Year']).mean().round(2)

#reset index
dfg = dfg.reset_index()
dfg.head()
```

```python
# define function to calculate price ratio for 2018 vs. 1998

def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [round(y2018/y1998, 2)]
    return ratio
```

```python
# create a dictionary with key-value pairs, borough:price ratio, by looping over all boroughs and calling the create_price_ratio function

# create empty dictionary
price_ratio = {}

# loop over all boroughs
for name in dfg['Borough'].unique():
    borough = dfg[dfg['Borough'] == name]
    
    # create entry in dictionary with name:price_ratio
    price_ratio[name] = create_price_ratio(borough)
```

```python
# create data frame, df_ratio, from dictionary

df_ratio = pd.DataFrame(price_ratio)
df_ratio.head()
```

```python
# transpose and reset index

df_ratio_fin = df_ratio.transpose()
df_ratio_fin = df_ratio_fin.reset_index() 
df_ratio_fin.head()
```

```python
# rename columns

df_ratio_fin.rename(columns={'index':'Borough', 0:'Price_ratio'}, inplace=True)
df_ratio_fin.head()
```

```python
# plot the increase of average house prices 2108 vs. 1998 for all boroughs 

plt.figure(figsize = (16, 8))

plt.bar(np.arange(1, len(df_ratio_fin)+1), df_ratio_fin['Price_ratio'], color = 'b')
plt.ylabel('Average House Price Increase', fontsize = 18, labelpad = 15)
plt.title('Average House Price Increase: 2018 vs. 1998', fontsize = 22, c = 'b', pad = 20)
plt.tick_params(labelsize = 15)
plt.show()
```

```python
# plot histogram of price ratios

plt.figure(figsize = (10, 10))

plt.hist(df_ratio_fin['Price_ratio'], bins = 8, color = 'b')
plt.xlabel('Price Ratio 2018 vs. 1998', fontsize = 18, labelpad = 15)
plt.ylabel('Frequency', fontsize = 18, labelpad = 15)
plt.title('Histogram of Price Ratio', fontsize = 22, c = 'b', pad = 20)
plt.tick_params(labelsize = 15)
plt.show()
```

```python
# get the average increase and the increase std

increase_avg = round(df_ratio_fin['Price_ratio'].mean(), 2)
increase_std = round(df_ratio_fin['Price_ratio'].std(), 2)

print('Average price increase from 1998 to 2018: ', increase_avg)
print('Standard deviation of price increase from 1998 to 2018: ', increase_std)
```

```python
# get the Top and Bottom 5 boroughs in average price increase

# Top 5
df_ratio_fin.sort_values(by='Price_ratio', ascending=False).head()
```

```python
# Bottom 5
df_ratio_fin.sort_values(by='Price_ratio', ascending=False).tail()
```
